```python
import control as ct
import numpy as np
import matplotlib.pyplot as plt

# Parámetros del sistema en espacio de estados
params = {
    "A": [[2, 5], [-4, -9]],  # Matriz de estado
    "B": [[3], [7]],          # Matriz de entrada
    "C": [[0, 6]],            # Matriz de salida
    "D": [[1]],               # Matriz de transmisión directa
    "num_puntos": 50         # Número de puntos en la simulación
}

# Crear el sistema en espacio de estados
sys = ct.ss(params["A"], params["B"], params["C"], params["D"])

# Obtener la respuesta al escalón
t, y = ct.step_response(sys, T=np.linspace(0, 10, params["num_puntos"]))

# Mostrar datos de entrada
print("Datos de Entrada: ")
print(f"Matriz A: {params['A']}")
print(f"Matriz B: {params['B']}")
print(f"Matriz C: {params['C']}")
print(f"Matriz D: {params['D']}")
print(f"Intervalo de tiempo: [0, 10]\n")

# Mostrar datos de salida (primeros 10 valores)
print("Datos de Salida: (Primeros 10 puntos)")
for i in range(10):
    print(f"Tiempo: {t[i]:.2f} s, Salida: {y[i]:.4f}")

# Personalización del color de la gráfica
color_grafica = "purple"  # Cambia el color aquí ('red', 'blue', etc.)

# Graficar la respuesta al escalón
plt.plot(t, y, color=color_grafica, label="Salida del sistema")
plt.title("Respuesta al Escalón de un Sistema en Espacio de Estados")
plt.xlabel("Tiempo (s)")
plt.ylabel("Salida")
plt.legend()
plt.grid()
plt.show()
```
